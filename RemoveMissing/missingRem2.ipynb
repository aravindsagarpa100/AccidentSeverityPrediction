```python
#convert string to int
df['Date'] = df['Date'].str.replace("-","").astype(int)
df['Time'] = df['Time'][df['Time'].notnull()].str.replace(":","").astype(int)
df.drop(df[df['Local_Authority_(Highway)'] == 'EHEATHROW'].index, inplace = True) 
df['Local_Authority_(Highway)'] = df['Local_Authority_(Highway)'][df['Local_Authority_(Highway)'].notnull()].str.replace('[A-Z]','').astype(int)
df['LSOA_of_Accident_Location'] = df['LSOA_of_Accident_Location'][df['LSOA_of_Accident_Location'].notnull()].str.replace('[A-Z]','').astype(int)
```

```python
#---[1]---
#mean part----replace mean by median/mode to fill by corresponding----
#df['Accident_Index'] = df['Accident_Index'].astype(int)
del df['Accident_Index']
for i in range(df.shape[1]):
    mean=df.iloc[:,i].mean()
    df.iloc[:,i] = df.iloc[:,i].fillna(mean)
```

```python
#---[2]---
#random state imputation
```

```python
cols = list(df.columns)
for i in range(df.shape[1]):
    #taking samples to replace NaN
    random_sample = df.iloc[:,i].dropna().sample(df.iloc[:,i].isnull().sum(),random_state=0)
    random_sample.index=df[df.iloc[:,i].isnull()].index
    df.loc[df.iloc[:,i].isnull(),cols[i]]=random_sample
```

```python
df['Date'] = df['Date'].str.replace("-","").astype(int)
df['Time'] = df['Time'][df['Time'].notnull()].str.replace(":","").astype(int)
df.drop(df[df['Local_Authority_(Highway)'] == 'EHEATHROW'].index, inplace = True) 
df['Local_Authority_(Highway)'] = df['Local_Authority_(Highway)'][df['Local_Authority_(Highway)'].notnull()].str.replace('[A-Z]','').astype(int)
df['LSOA_of_Accident_Location'] = df['LSOA_of_Accident_Location'][df['LSOA_of_Accident_Location'].notnull()].str.replace('[A-Z]','').astype(int)
```

```python
del df['Accident_Index']
for i in range(df.shape[1]):
    extreme = df.iloc[:,i].mean() + 3 * df.iloc[:,i].std()
    df.iloc[:,i] = df.iloc[:,i].fillna(extreme)
```
